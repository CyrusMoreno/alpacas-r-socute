```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('202211-divvy-tripdata.csv')

# Now, 'df' is a DataFrame object containing the data from your CSV file
```

```python
df.head()
```

```python
# Using nunique()
unique_count = df['start_station_name'].nunique()
print(f"Number of unique values: {unique_count}")

# Using value_counts()
value_counts = df['start_station_name'].value_counts()
print("Counts of unique values:")
print(value_counts)
```

```python
# Assuming df is your DataFrame and 'column_name' is the name of your column
empty_cells = df['ride_id'].isnull().sum()
print(f"Number of empty cells: {empty_cells}")
```

```python
# create new columns for starting coordinates and destination column
df['starting_coords'] = 'coordinates: ' + df['start_lat'].astype(str) + ' , ' + df['start_lng'].astype(str)
df['destination_coords']= 'coordinates: ' + df['end_lat'].astype(str) + ' , ' + df['end_lng'].astype(str)
df.head()
```

```python
#fill empty statiion names with coordinates
df['start_station_name'] = df['start_station_name'].fillna(df['starting_coords'])
df['end_station_name'] = df['end_station_name'].fillna(df['destination_coords'])

ride_with_no_start_station_names = df['start_station_name'].isnull().sum()
print(f"Number of no start station name: {ride_with_no_start_station_names}")

ride_with_no_end_station_names = df['start_station_name'].isnull().sum()
print(f"Number of no end station name: {ride_with_no_end_station_names}")
```

```python
# Convert the start_time and end_time columns to datetime
df['started_at'] = pd.to_datetime(df['started_at'], format='%Y-%m-%d %H:%M:%S')
df['ended_at'] = pd.to_datetime(df['ended_at'], format='%Y-%m-%d %H:%M:%S')

# Calculate the duration and store it in a new column
df['duration'] = df['ended_at'] - df['started_at']

df.head()
```
